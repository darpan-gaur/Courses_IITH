## Problem 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sketch of function f(\phi) = \lambda(\phi^2-1)^2
# it's first derivative is f'(\phi) = 4\lambda\phi(\phi^2-1)
# it's second derivative is f''(\phi) = 4\lambda(3\phi^2-1)
def f(phi, lambda_val):
    return lambda_val*(phi**2-1)**2

def f_prime(phi, lambda_val):
    return 4*lambda_val*phi*(phi**2-1)

def f_double_prime(phi, lambda_val):
    return 4*lambda_val*(3*phi**2-1)

phi = np.linspace(-1.5, 1.5, 1000)
lambda_val = 1
plt.plot(phi, f(phi, lambda_val), label=r'$f(\phi) = \lambda(\phi^2-1)^2$')
plt.plot(phi, f_prime(phi, lambda_val), label=r"$f'(\phi) = 4\lambda\phi(\phi^2-1)$")
plt.plot(phi, f_double_prime(phi, lambda_val), label=r"$f''(\phi) = 4\lambda(3\phi^2-1)$")
plt.legend()
plt.xlabel(r'$\phi$')
plt.ylabel('f')
plt.tight_layout()
plt.show()
# plt.savefig('function_sketch_all.png')

plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
plt.plot(phi, f(phi, lambda_val), label=r'$f(\phi) = \lambda(\phi^2-1)^2$')
plt.xlabel(r'$\phi$')
plt.ylabel('f')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(phi, f_prime(phi, lambda_val), label=r"$f'(\phi) = 4\lambda\phi(\phi^2-1)$")
plt.xlabel(r'$\phi$')
plt.ylabel('f')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(phi, f_double_prime(phi, lambda_val), label=r"$f''(\phi) = 4\lambda(3\phi^2-1)$")
plt.xlabel(r'$\phi$')
plt.ylabel('f')
plt.legend()

plt.tight_layout()
plt.show()
# plt.savefig('function_sketch.png')



```

```python
import sympy as sp
```

### Part (a)

```python
# Define symbols and function
x = sp.symbols('x')
phi = sp.Function('phi')(x)
phi_prime = sp.Derivative(phi, x)
f_phi = sp.Function('f')(phi)
k = sp.symbols('k')

# define the lagrangian
F = f_phi + (1/2)*k*phi_prime**2 

# compute partial derivatives
dF_dphi = sp.diff(F, phi)
dF_dphi_prime = sp.diff(F, phi_prime)

# compute variational derivative
dF_dphi_prime_derivative = sp.diff(dF_dphi_prime, x)
variational_derivative = dF_dphi - dF_dphi_prime_derivative

# display the result
sp.simplify(variational_derivative)
```
