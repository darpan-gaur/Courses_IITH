```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# take out.txt as input and split intto a list
with open('out.txt', 'r') as f:
    data = f.read().split('\n')

data = data[1:]
data = [i.split(' ') for i in data]

# take average of data[i][1] for 6 items in data
avgC = []
avgM = []
tC = []
tM = []
for i in range(0, len(data), 6):
    if (len(data[i]) < 3):
        continue
    if (i%4==0):
        # print(i, "CLH", data[i])
        avgC.append(np.mean([float(j[1]) for j in data[i:i+5]]))
        tC.append(np.mean([float(j[2]) for j in data[i:i+5]]))
    else:
        # print(i, "MCS", data[i])
        avgM.append(np.mean([float(j[1]) for j in data[i:i+5]]))
        tM.append(np.mean([float(j[2]) for j in data[i:i+5]]))
```

```python
print(avgC)
avgC = 1e3*np.array(avgC)
print(avgM)
avgM = 1e3*np.array(avgM)
print(tC)
print(tM)
```

```python
# plot the graph
plt.plot([2, 4, 8, 16, 32, 64], avgC, label='CLH', marker='o')
# triangle marker
plt.plot([2, 4, 8, 16, 32, 64], avgM, label='MCS', marker='^')
plt.grid()
plt.xlabel('Number of Threads')
plt.ylabel('Average Entry Time (ms)')
plt.title('Average Entry Time vs Number of Threads')
plt.legend()
```

```python
# plot the graph
plt.plot([2, 4, 8, 16, 32, 64], tC, label='CLH', marker='o')
plt.plot([2, 4, 8, 16, 32, 64], tM, label='MCS', marker='^')
plt.grid()
plt.xlabel('Number of Threads')
plt.ylabel('Throughput (ops/sec)')
plt.title('Throughput vs Number of Threads')
plt.legend()
```
