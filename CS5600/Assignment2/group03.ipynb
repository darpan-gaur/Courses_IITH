```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from spmf2 import Spmf
```

```python
input_file1 = './dataset/retail1/retail1.txt'
output_file1 = './output/retail1.txt'

input_file2 = './dataset/retail2.txt'
output_file2 = './output/retail2.txt'
```

## Exercise 1

```python
algorithm = ['Apriori', 'FPGrowth_itemsets', 'Eclat']
min_sup = [0.005, 0.01, 0.02, 0.03, 0.05, 0.07]
```

```python
time_list = []
for algo in algorithm:
    time_temp = []
    for sup in min_sup:
        spmf = Spmf(algorithm_name=algo, input_filename=input_file1,
                    output_filename=output_file1, arguments=[sup])
        st = spmf.run()
        st = st.split('\n')
        # time_taken = st[-3].split(' ')[-2] # time taken
        # search for 'Total time' in the output
        for i in range(len(st)):
            if 'Total time' in st[i]:
                time_taken = int(st[i].split(' ')[-2])
                break
        time_temp.append(time_taken)
    time_list.append(time_temp)
        
```

```python
time_list
```

```python
# plot the time taken for each algorithm
fig, ax = plt.subplots()
ax.plot(min_sup, time_list[0], label='Apriori', marker='o')
ax.plot(min_sup, time_list[1], label='FPGrowth', marker='o')
ax.plot(min_sup, time_list[2], label='Eclat', marker='o')
ax.set_xlabel('Minimum Support')
ax.set_ylabel('Time taken (ms)')
ax.set_title('Time taken for different algorithms on retial1.txt')
ax.legend()
plt.show()
```

```python
# min_sup = 0.5%, ls = 0.01
# use fpgrowth_itemsets and print the itemsets
spmf = Spmf(algorithm_name='FPGrowth_itemsets', input_filename=input_file1,
            output_filename=output_file1, arguments=[0.005])
st = spmf.run()
```

```python
# get top 5 itemsets, with the highest support, from the output
# read the output file
with open(output_file1, 'r') as f:
    lines = f.readlines()

# get the itemsets which r of type '1909 2204 #SUP: 3052', left of #SUP is the itemset and right is the support
itemsets = []
for line in lines:
    line = line.split()
    if line[-2] == '#SUP:':
        itemset = line[:-2]
        support = int(line[-1])
        if (len(itemset) > 1):
            itemsets.append((itemset, support))

# sort the itemsets based on support
itemsets = sorted(itemsets, key=lambda x: x[1], reverse=True)

# print the top 5 itemsets
for i in range(5):
    print(itemsets[i])
```

```python
cannot_be_together = itemsets[:5]
cannot_be_together = [x[0] for x in cannot_be_together]
print(f'cannot_be_together = {cannot_be_together}')
```

```python
# take the item with least item id, and union with all other items from cannot_be_together
# this will be the must_have
must_have = [min(int(x) for x in cannot_be_together[i]) for i in range(5)]
# sort the must_have
must_have = set(must_have)
print(f'must_have = {must_have}')
```
