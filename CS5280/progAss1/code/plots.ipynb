```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Plot 1: Number of Transactions

```python
# bocc
# 1000 1.56546e+06 2.894
# 2000 1.57749e+06 2.924
# 3000 1.58333e+06 2.93567
# 4000 1.59501e+06 2.96425
# 5000 1.61081e+06 3.001

# focc
# 5000 -> 4.74553e+06 10.8106
# 4000 -> 4.81234e+06 10.963
# 3000 -> 4.93361e+06 11.2823
# 2000 -> 4.33997e+06 9.803
# 1000 -> 4.84653e+06 11.0616
```

```python
numTrans = np.array([1000, 2000, 3000, 4000, 5000])
bocc_totalDealy = np.array([1.56546e+06, 1.57749e+06, 1.58333e+06, 1.59501e+06, 1.61081e+06])
bocc_aborts = np.array([2.894, 2.924, 2.93567, 2.96425, 3.001])
focc_totalDealy = np.array([4.84653e+06, 4.33997e+06, 4.93361e+06, 4.81234e+06, 4.74553e+06])
focc_aborts = np.array([11.0616, 9.803, 11.2823, 10.963, 10.8106])

plt.plot(numTrans, bocc_totalDealy, 'r', label='BOCC', marker='o')
plt.plot(numTrans, focc_totalDealy, 'b', label='FOCC', marker='o')
plt.xlabel('Number of Transactions')
plt.ylabel('commit delay (microseconds)')
plt.legend()
plt.title('Commit Delay vs Number of Transactions')
plt.grid()
plt.savefig('commitDelayVsNumTrans.png')
plt.show()

plt.plot(numTrans, bocc_aborts, 'r', label='BOCC', marker='o')
plt.plot(numTrans, focc_aborts, 'b', label='FOCC', marker='o')
plt.xlabel('Number of Transactions')
plt.ylabel('NUmber of aborts')
plt.legend()
plt.title('Number of aborts vs Number of Transactions')
plt.grid()
plt.savefig('numAbortsVsNumTrans.png')
plt.show()
```
