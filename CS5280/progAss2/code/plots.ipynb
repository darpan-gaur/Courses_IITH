```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getData(lines, param):
    btoTD, btoAC = [], []
    mvtoTD, mvtoAC = [], []
    mvtogcTD, mvtogcAC = [], []
    kmvtoTD, kmvtoAC = [], []
    i=0
    while i < len(lines):
        l = lines[i]
        if 'BTO' in l:
            i+=1
            TD = []
            AC = []
            while i < len(lines) and param not in lines[i]:
                l = lines[i].split()
            # print(l)
                TD.append(float(l[0]))
                AC.append(float(l[1]))
                i+=1
        # print(TD)
        # print(AC)
            TD = np.mean(np.array(TD))
            btoTD.append(np.mean(np.array(TD)))
            btoAC.append(np.mean(np.array(AC)))
        elif 'MVTO-gc' in l:
            i+=1
            TD = []
            AC = []
            while i < len(lines) and param not in lines[i]:
                l = lines[i].split()
            # print(l)
                TD.append(float(l[0]))
                AC.append(float(l[1]))
                i+=1
        # print(TD)
        # print(AC)
            TD = np.mean(np.array(TD))
            mvtogcTD.append(np.mean(np.array(TD)))
            mvtogcAC.append(np.mean(np.array(AC)))
        elif 'k-MVTO' in l:
            i+=1
            TD = []
            AC = []
            while i < len(lines) and param not in lines[i]:
                l = lines[i].split()
            # print(l)
                TD.append(float(l[0]))
                AC.append(float(l[1]))
                i+=1
        # print(TD)
        # print(AC)
            TD = np.mean(np.array(TD))
            kmvtoTD.append(np.mean(np.array(TD)))
            kmvtoAC.append(np.mean(np.array(AC)))
        elif 'MVTO' in l:
            i+=1
            TD = []
            AC = []
            while i < len(lines) and param not in lines[i]:
                l = lines[i].split()
            # print(l)
                TD.append(float(l[0]))
                AC.append(float(l[1]))
                i+=1
        # print(TD)
        # print(AC)
            TD = np.mean(np.array(TD))
            mvtoTD.append(np.mean(np.array(TD)))
            mvtoAC.append(np.mean(np.array(AC)))
    
    
        else:
            i+=1
    btoTD = np.array(btoTD)
    mvtoTD = np.array(mvtoTD)
    mvtogcTD = np.array(mvtogcTD)
    kmvtoTD = np.array(kmvtoTD)
    btoAC = np.array(btoAC)
    mvtoAC = np.array(mvtoAC)
    mvtogcAC = np.array(mvtogcAC)
    kmvtoAC = np.array(kmvtoAC)

    return btoTD, btoAC, mvtoTD, mvtoAC, mvtogcTD, mvtogcAC, kmvtoTD, kmvtoAC
```
