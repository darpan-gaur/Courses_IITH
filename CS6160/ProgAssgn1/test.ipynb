```python
# input 12 cyphertexts in hex from input file and store in list
ciphertexts = []
with open('streamciphertexts.txt', 'r') as f:
    for line in f:
        ciphertexts.append(line.strip())

for i in range(len(ciphertexts)):
    print(f'ciphertext {i+1}: {ciphertexts[i]}')
```

```python
# convert hex ciphertexts to decimal by taking 2 hex characters at a time and converting to decimal
ciphertexts_dec = []
for ciphertext in ciphertexts:
    ciphertexts_dec.append([int(ciphertext[i:i+2], 16) for i in range(0, len(ciphertext), 2)])

# print decimal ciphertexts
for i in range(len(ciphertexts_dec)):
    print(f'ciphertext {i+1}: {ciphertexts_dec[i]}')  
```

```python
keySz = max(len(ciphertext) for ciphertext in ciphertexts_dec)

Ck = [[] for i in range(keySz)]

# make list of ith characters of each ciphertext
for ciphertext in ciphertexts_dec:
    for i, item in enumerate(ciphertext,0):
        tempK = Ck.pop(i)
        tempK.append(item)
        Ck.insert(i,tempK)

for i in Ck:
    print(i)
```

```python
messages = list()
for i in ciphertexts_dec:
    messages.append(['?']*len(i))

for i in messages:
    print(i)
    
```

```python
key = list()

for k in Ck:
    freqTable = {}
    spaceAtK = -1
    for i in k:
        for j in k:
            result = i ^ j
            if (result >= 65):
                if i not in freqTable:
                    freqTable[i] = 1
                else:
                    freqTable[i] = freqTable.get(i) + 1
                if j not in freqTable:
                    freqTable[j] = 1
                else:
                    freqTable[j] = freqTable.get(j) + 1
                spaceAtK = max(freqTable, key=freqTable.get)
    if (spaceAtK == -1):
        key.append(-1)
    else:
        key.append(spaceAtK ^ 32)

print(key)
    
```

```python
# find message using taking xor of key and ciphertext
for i in range(len(ciphertexts_dec)):
    for j in range(len(ciphertexts_dec[i])):
        if (key[j] == -1):
            messages[i][j] = 63
        else:
            messages[i][j] = ciphertexts_dec[i][j] ^ key[j]
```

```python
# print clear text by converting decimal to ascii
for message in messages:
    print(f'M{messages.index(message)+1}: ', end='')
    print(''.join(chr(i) for i in message))
```
